```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
data_df = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')

# Display the data table for preview
data_df.head()
```

```python
# Checking the number of mice.
unique_mice = len(data_df["Mouse ID"].unique())
unique_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# Mouse ID: g989 appears 13 times which appears to be an outlier of the other Mouse IDs that comes up 10 times.
duplicate = data_df.loc[data_df.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
duplicate
```

```python
# Optional: Get all the data for the duplicate mouse ID.
duplicate_df = data_df.loc[(data_df['Mouse ID'] == 'g989')] 
duplicate_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data_df = data_df.loc[(data_df['Mouse ID'] != 'g989')] 
clean_data_df 
```

```python
# Checking the number of mice in the clean DataFrame.
clean_unique_mice = len(clean_data_df["Mouse ID"].unique())
clean_unique_mice
```

# Summary Statistics
Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

```python
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
drug_df = clean_data_df.groupby("Drug Regimen")

# mean, median, variance, standard deviation, and SEM of the tumor volume.
tumor_mean = drug_df['Tumor Volume (mm3)'].mean()
tumor_median = drug_df['Tumor Volume (mm3)'].median()
tumor_std = drug_df['Tumor Volume (mm3)'].std()
tumor_var = drug_df['Tumor Volume (mm3)'].var()
tumor_sem = drug_df['Tumor Volume (mm3)'].sem()

# Assemble the resulting series into a single summary DataFrame.
tumor_df = pd.DataFrame({
                        "Mean Tumor Volume": tumor_mean,
                        "Median Tumor Volume" : tumor_median,
                        "Tumor Volume Variance " : tumor_var,
                        "Tumor Volume Std. Dev." : tumor_std,
                        "Tumor Volume Std. Err." : tumor_sem
                        })
tumor_df.reset_index()
```

```python
# Using the aggregation method, produce the same summary statistics in a single line.
tumor_agg_df = drug_df.agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
tumor_agg_df
```

# Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
max_timepoint = clean_data_df.groupby("Mouse ID")["Timepoint"].max().sort_values(ascending=False)
max_timepoint_df = pd.DataFrame(max_timepoint)
max_timepoint_df.head()
```

```python
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
max_df = pd.merge(max_timepoint_df, clean_data_df, on = ["Mouse ID", "Timepoint"])
max_df.reset_index()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Locate the rows which contain mice on each drug and get the tumor volumes
for treatment in treatments:
    tumor_volume = max_df.loc[max_df["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]
    
    # add subset 
    tumor_vol_data.append(tumor_volume)
    
    # Determine outliers using upper and lower bounds
    quartiles = tumor_volume.quantile([0.25, 0.5, 0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    print(f"IQR for {treatment} is {iqr}")
    
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"Lower bound for {treatment} is {lower_bound}")
    print(f"Upper bound for {treatment} is {upper_bound}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
# Outlier
display = dict(marker='o', markerfacecolor ='r', markersize = 10, markeredgecolor ='black')

# Plot
plt.boxplot(tumor_vol_data, flierprops = display);

# Labelling
plt.title("Final Tumor Volume Vs. Drug Regimen")
plt.xticks([1,2,3,4],["Capomulin", "Ramicane", "Infubinol", "Ceftamin"])
plt.xlabel("Drug Regimen")
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

# Line and Scattor Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
# Mouse treated with Capomulin
capomulin_df = clean_data_df.loc[clean_data_df["Drug Regimen"] == "Capomulin"]

# Data specific for Mouse l509
mouse_l509 = capomulin_df.loc[capomulin_df["Mouse ID"] == "l509"]

# Plotting
plt.plot(mouse_l509["Timepoint"], mouse_l509["Tumor Volume (mm3)"])
plt.title("Mouse l509's Tumor Volume Over Time")
plt.xlabel("Timepoint (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
avg_vol = capomulin_df.groupby(["Mouse ID"]).mean(numeric_only=True)

# Scatter graph
plt.scatter(avg_vol["Weight (g)"], avg_vol["Tumor Volume (mm3)"])
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

# Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

# Corellation Coefficient
correlation = st.pearsonr(avg_vol["Weight (g)"], avg_vol["Tumor Volume (mm3)"])

# x-axis and y-axis
x = avg_vol["Weight (g)"]
y = avg_vol["Tumor Volume (mm3)"]

# Values for a line
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x,y)

# Regression values
regress = x * slope + intercept
line_eq = "y = " + str(round(slope, 2)) + "x + " + str(round(intercept, 2))
plt.scatter(x,y)
plt.plot(x, regress, "r-")

# Labelling
plt.annotate(line_eq, (20,37), fontsize = 15, color = "red")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()

print(f"The r-squared value is {rvalue**2}")
```
